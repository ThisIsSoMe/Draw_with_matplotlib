```python
import matplotlib.pyplot as plt
import numpy as np

# edgecolor='#4D79C8'
# 图的数据
data=[[78,11,15,1,2],[89,8,2,0,3],[22,65,13,43,1]]
labels = ['matching','calculation','sorting','clustering','others']
legends = ['Information retrieval','Customer service','Data analysis']
hatchs = ['','///','---']
colors=['deepskyblue','orange','sandybrown']

# 坐标轴和刻度位置
x = np.arange(len(data[0]))
total_width, n = 0.8, len(data)
width = total_width / n
x = x - (total_width - width) / 2
fig, ax = plt.subplots()

# 画图
for i in range(len(data)):
    # 画直方体
    ax.bar(x+i*width, data[i],width=width,color=colors[i],label=legends[i],tick_label=labels,hatch=hatchs[i],edgecolor='black')
    # 添上数据
    for ind, j in enumerate(data[i]):
        ax.text(ind+(i-1)*width, j+2, j, ha='center',fontsize=10)
ax.set_yticks(np.arange(0,101,20))
ax.set_xlabel('Question type')
ax.set_ylabel('Proportion (%)')

# legend and show
plt.legend(loc = 'best')
fig.show()
```
